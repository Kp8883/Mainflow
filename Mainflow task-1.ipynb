```python
#list
list=[1,2,3,4]
print("Original List\n",list)


list.append(6)
print("Appended List\n",list)

list.extend([97,8,21,23])
print("Extended List\n",list)

list.insert(5,45)
print("Inserted List\n",list)

list.remove(3)
print("Removing an element from List\n",list)
list.pop()
print("Popping in List\n",list)

list[2]=11
print("Changing element in List\n",list)

list.reverse()
print("Reversing the List\n",list)

print("Slicing(Showing only specific elements of List)\n",list[2:5])

print("Count the number of total elements in List\n",len(list))
print("Returns maximum element of List\n",max(list))
print("Returns minimum element of List\n",min(list))

list.extend([90,20,20,80,2,2,2,5])
print(list)

print("Count the number of copies in List\n",list.count(2),list.count(20),list.count(100),list.count(90))

list2=[23,43,44,46,56,79,80]
print(list2)

new_list= list+list2
print("Concatination of list\n",new_list)

multiply_list=list*2
print("Writes the elements multiple times in list\n",multiply_list)

list.sort()
print("Sorted List\n",(list))

list.clear()
print("Returns empty list\n",list)
```

```python
#dictionary
dic={'name':'Rohan','age':20,'city':'Mumbai','profession':'student'}
print(dic)

dic['gender']='male'
print(dic)

del dic['age']
print(dic)

dic['city']='Bangalore'
print(dic)

print(dic.get('age'))
print(dic.get('city'))

from typing import List
print(list(dic.items())[2][0])
print(list(dic.items())[2][1])
print(list(dic.keys()))

dic1={'name':'Rahul','age':30,'city':'Delhi'}
dic.update(dic1)
print(dic)
     
dic.clear()
print(dic)
```

```python
set={1,2,3,4}
print("Elements of set\n",set)

set2={23,3,2,78,90,12}
print("Elements of set2\n",set2)

union_set=set | set2
print("Union of set2\n",union_set)

intersection_set=set | set2
print("Intersection of set2\n",intersection_set)

print(set.issubset(set2))
print(set2.issubset(set))
setn={1,2}
print(setn.issubset(set))
print(setn.issubset(set2))

print(set.issuperset(set2))
print(set2.issuperset(set))
setn={1,2}
print(set.issuperset(setn))
print(set2.issuperset(setn))

set3= set-set2
print("Difference of sets",set3)

set.add(6)
print(set)

set.remove(3)
print(set)

set.discard(2)
print(set)


```
